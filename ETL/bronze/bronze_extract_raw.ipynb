# DEFINE EXTRACTION CONFIGURATION

```python
# SOURCE
INGESTION_CONFIG = [
    {
        "source": "assets",
        "path": "/Volumes/fm_prj1/source/data/assets.csv",
        "table": "assets"
    },
    {
        "source": "technicians",
        "path": "/Volumes/fm_prj1/source/data/technicians.csv",
        "table": "technicians"
    },
    {
        "source": "work_orders",
        "path": "/Volumes/fm_prj1/source/data/work_orders.csv",
        "table": "work_orders"
    },
]

# DESTINATION
CATALOG = "fm_prj1"
SCHEMA = "bronze"
```

# EXTRACT FILES INTO BRONZE TABLES

```python
for item in INGESTION_CONFIG:
    print(f"Ingesting {item['source']} → {CATALOG}.{SCHEMA}.{item['table']}")

    df = (
        spark.read
            .option("header", "true")
            .option("inferSchema", "false")
            .csv(item["path"])
    )

    (
        df.write
            .format("delta")
            .mode("overwrite")
            .saveAsTable(f"{CATALOG}.{SCHEMA}.{item['table']}")
    )

print("Ingested raw data to bronze schema successfully!")
```
